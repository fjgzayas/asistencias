```
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

# Check if the image file exists
image_path = '/workspace/uploads/image.png'
if os.path.exists(image_path):
    # Load and display the image
    img = mpimg.imread(image_path)
    
    # Create a figure to display the image
    plt.figure(figsize=(12, 8))
    plt.imshow(img)
    plt.axis('off')
    plt.title('Bug Report Image - Task Management Application')
    plt.tight_layout()
    plt.show()
    
    print("Image loaded successfully!")
    print(f"Image shape: {img.shape}")
    print("Analyzing the bug shown in the image...")
else:
    print(f"Image file not found at: {image_path}")
    print("Let me check what files are available in the uploads directory:")
    
    uploads_dir = '/workspace/uploads'
    if os.path.exists(uploads_dir):
        files = os.listdir(uploads_dir)
        print(f"Files in uploads directory: {files}")
    else:
        print("Uploads directory does not exist")
```
